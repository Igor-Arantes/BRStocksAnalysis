```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import  Image
%matplotlib inline
```

```
df = pd.read_csv('iris.csv')
```

```
fig = plt.figure(figsize=(15,8))
axis = fig.add_axes([0,0,1,1])

axis.scatter(df['comprimento_sépala'], df['largura_sépala'])
axis.set_title('Grafico de dispersão', fontsize =25,pad=15)
axis.set_xlabel('Comprimento da sépala',fontsize=15)
axis.set_ylabel('Largura da sépala',fontsize=15)

plt.show()
```

```
fig = plt.figure(figsize=(15,8))
axis = fig.add_axes([0,0,1,1])

axis.scatter(df['comprimento_sépala'], df['largura_sépala'])
axis.set_title('Grafico de dispersão', fontsize =25,pad=15)
axis.set_xlabel('Comprimento da sépala',fontsize=15)
axis.set_ylabel('Largura da sépala',fontsize=15)

axis.tick_params(labelsize=15)


plt.show()
```

```
fig = plt.figure(figsize=(15,8))
axis = fig.add_axes([0,0,1,1])

colors={'Iris-setosa':'r','Iris-versicolor':'b', 'Iris-virginica':'g' }
markers={'Iris-setosa':'o','Iris-versicolor':'x', 'Iris-virginica':'v' }

for species in df['espécie'].unique():
  temp = df[df['espécie']== species]
  axis.scatter(temp['comprimento_sépala'], temp['largura_sépala'], color=colors[species]
               , marker=markers[species], s = 100)


#axis.scatter(df['comprimento_sépala'], df['largura_sépala'])
axis.set_title('Grafico de dispersão', fontsize =25,pad=15)
axis.set_xlabel('Comprimento da sépala',fontsize=15)
axis.set_ylabel('Largura da sépala',fontsize=15)

axis.tick_params(labelsize=15)

axis.legend(colors, fontsize = 20)

plt.show()
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

axis.boxplot(df['largura_pétala'])
axis.set_title('Grafico de caixa',fontsize=15, pad=10)
axis.set_xticklabels(df['largura_pétala'])
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

axis.boxplot(df.drop('espécie',axis=1).values)
axis.set_title('Grafico de caixa',fontsize=15, pad=10)
axis.set_xticklabels(df.drop('espécie',axis=1).columns)
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

colors=['red','blue','orange','green']

boxes = axis.boxplot(df.drop('espécie',axis=1).values, patch_artist=True)
axis.set_title('Grafico de caixa',fontsize=15, pad=10)
axis.set_xticklabels(df.drop('espécie',axis=1).columns)

for box, color in zip(boxes['boxes'],colors):
  box.set(color = color)

for outlier in boxes['fliers']:
  outlier.set(marker='x',markersize=10)
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

axis.hist(df['comprimento_pétala'])
axis.set_title('Histograma', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

axis.hist(df['comprimento_pétala'],bins=20)
axis.set_title('Histograma', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

axis.hist(df['comprimento_pétala'],bins=20, density=True)
axis.set_title('Histograma', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()
axis.hist(df['comprimento_pétala'],bins=20)
axis.set_title('Histograma', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)

axis.annotate('$mu = {0:.2f}$\n$sigma = {0:.2f}$'.format(mu, sigma),xy=(4.5,20), fontsize=20)
```

```
fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()
axis.hist(df['comprimento_pétala'],bins=20)
axis.set_title('Histograma', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)

axis.annotate('$\mu = {0:.2f}$\n$\sigma = {0:.2f}$'.format(mu, sigma),xy=(4.5,20), fontsize=20)

axis.axvline(mu,color = 'k',linestyle='--')
axis.annotate('Média',xy=(mu-1, 28), fontsize=20)


axis.axvline(df['comprimento_pétala'].median(),color = 'g',linestyle='--')
axis.annotate('Mediana',xy=(df['comprimento_pétala'].median()+0.2, 28), fontsize=20)
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

df_iv=df[df['espécie']== 'Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()
axis.hist(df_iv['comprimento_pétala'],bins=20)
axis.set_title('Iris-Versicolor', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)

axis.annotate('$\mu = {0:.2f}$\n$\sigma = {0:.2f}$'.format(mu, sigma),xy=(4.6,6), fontsize=20)

axis.axvline(mu,color = 'k',linestyle='--')
axis.annotate('Média',xy=(mu-0.3, 5.5), fontsize=20)


axis.axvline(df_iv['comprimento_pétala'].median(),color = 'g',linestyle='--')
axis.annotate('Mediana',xy=(df_iv['comprimento_pétala'].median()-0.45, 6),color='g', fontsize=20)

fig.savefig('Hisograma_iv.png', bbox_inches='tight')
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

df_is=df[df['espécie']== 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()
axis.hist(df_is['comprimento_pétala'],bins=20)
axis.set_title('Iris-Setosa', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)

axis.annotate('$\mu = {0:.2f}$\n$\sigma = {0:.2f}$'.format(mu, sigma),xy=(1.7,7), fontsize=20)

axis.axvline(mu,color = 'k',linestyle='--')
axis.annotate('Média',xy=(mu-0.2, 13.5), fontsize=20)


axis.axvline(df_is['comprimento_pétala'].median(),color = 'g',linestyle='--')
axis.annotate('Mediana',xy=(df_is['comprimento_pétala'].median()+0.05, 12),color='g', fontsize=20)

fig.savefig('Hisograma_is.png', bbox_inches='tight')
```

```
fig = plt.figure(figsize=(5,4))
axis = fig.add_axes([0,0,1,1])

df_ivc=df[df['espécie']== 'Iris-virginica']

mu, sigma = df_ivc['comprimento_pétala'].mean(), df_ivc['comprimento_pétala'].std()
axis.hist(df_ivc['comprimento_pétala'],bins=20)
axis.set_title('Iris-virginica', fontsize=15 , pad=10)
axis.set_xlabel('Comprimento da pétala', fontsize=15)
axis.grid(True)

axis.annotate('$\mu = {0:.2f}$\n$\sigma = {0:.2f}$'.format(mu, sigma),xy=(6.3,7), fontsize=20)

axis.axvline(mu,color = 'k',linestyle='--')
axis.annotate('Média',xy=(mu, 8), fontsize=20)


axis.axvline(df_ivc['comprimento_pétala'].median(),color = 'g',linestyle='--')
axis.annotate('Mediana',xy=(df_ivc['comprimento_pétala'].median(), 7),color='g', fontsize=20)

fig.savefig('Hisograma_ivc.png', bbox_inches='tight')
```

```
width, heigh = Image.open('Hisograma_ivc.png').size
combine = Image.new('RGB',(3*width,heigh))
interval = 0
for image in map(Image.open, ['Hisograma_iv.png','Hisograma_is.png','Hisograma_ivc.png']):
  combine.paste(image, (interval,0))
  interval += width

combine.save('Combined Histograms.png')
```
